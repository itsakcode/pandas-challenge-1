## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
client_df['sub_total'] = (client_df['qty'] * client_df['unit_price'])
client_df
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

def get_shipping_price(x):
    """
        This function is to calculate shipping price.

        Optional:
        use unit_weight, qty to calculate the weight and then 
        shipping price.

        Args:
        Row of the dataframe

        Returns:
        Float value which is calculated shipping price
    """
    total_weight = (x['unit_weight'] * x['qty'])
    
    if total_weight <= 50:
        return round((total_weight * 10),2)
    return round((total_weight * 7),2)

client_df['shipping_price'] = client_df.apply(get_shipping_price, axis=1)
client_df
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
client_df['total_price'] = round(((client_df['sub_total'] + \
                                   client_df['shipping_price']) * 1.0925), 2)
client_df
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

#it can also be done using function. 
""" def calculate_cost(x):
    return ((x['unit_cost'] * x['qty']) + x['shipping_price'])

client_df['total_cost'] = client_df.apply(calculate_cost, axis=1) """

#can also be implemented using lamda
client_df['total_cost'] = client_df.\
    apply(lambda x: ((x['unit_cost'] * x['qty']) + x['shipping_price']), axis=1)
client_df
```

```python
# Create a column for the profit of
# each line using line cost and line price

client_df['profit'] = client_df.apply(lambda i: (i['total_price'] - \
                                                 i['total_cost']), axis=1)
client_df
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
top_spenders = {}

# Sort the DataFrame by total_price in descending order and get top 5 client_ids
for i in top_clients:
    top_spenders[i] = round((client_df['total_price']\
                             .loc[(client_df['client_id'] == i)].sum()), 2)
    
print("Top 5 Client IDs:", top_spenders)
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

# Extract the rows with just top client ids
top_5_clients_df = client_df.loc[(client_df['client_id'].isin(top_clients))]
top_5_clients_df

# create lists of data for each column you want in summary
top_5_clients = []
top_5_total_units = []
top_5_total_shipping_price = []
top_5_total_revenue = []
top_5_total_profits = []

# populate the lists in same order of client id so you can make sure 
# the data is correct
for i in top_clients:
    top_5_clients.append(i)
    top_5_total_units.append(client_df['qty']\
                             .loc[(client_df['client_id'] == i)].sum())
    top_5_total_shipping_price.append(round((client_df['shipping_price']\
                             .loc[(client_df['client_id'] == i)].sum()), 2))
    top_5_total_revenue.append(top_spenders[i])
    top_5_total_profits.append(round((client_df['profit']\
                             .loc[(client_df['client_id'] == i)].sum()), 2))

# use above lists to create the summary dataframe
top_clients_summary_df = \
    pd.DataFrame({'client_id': top_5_clients, \
                    'total_units': top_5_total_units, \
                    'total_shipping': top_5_total_shipping_price, \
                    'total_revenue': top_5_total_revenue, \
                    'total_profit': top_5_total_profits
                }).sort_values(by='total_profit')

#top_clients_summary_df.dtypes
top_clients_summary_df
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

# convert all currency values to millions
top_clients_summary_df[['total_shipping', 'total_revenue', 'total_profit']] = \
    top_clients_summary_df[['total_shipping', 'total_revenue', 'total_profit']] \
        .apply(lambda x: round(x/1000000,2))

# rename the columns
top_clients_summary_df.rename(\
    columns={'client_id': 'Client ID', \
                'total_units': 'Total Units', \
                'total_shipping': 'Total Shipping (in Millions)', \
                'total_revenue': 'Total Revenue (in Millions)', \
                'total_profit': 'Total Profit (in Millions)'}, inplace=True)


top_clients_summary_df
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
top_clients_summary_sorted_df = top_clients_summary_df.sort_values('Total Profit (in Millions)', ascending=False)
top_clients_summary_sorted_df.reset_index(drop=True, inplace=True)
top_clients_summary_sorted_df

# write the updated and sorted dataframe to csv
#top_clients_summary_sorted_df.to_csv("Output/top_clients_summary.csv")
```
